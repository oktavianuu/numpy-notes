```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a 3D array with shape (3, 3, 3) as an example
array_3d = np.random.rand(3, 3, 3)

# Set up a 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot each point in the 3D array
for i in range(array_3d.shape[0]):
    for j in range(array_3d.shape[1]):
        for k in range(array_3d.shape[2]):
            # Plot each element as a point in 3D space
            ax.scatter(i, j, k, color="b")

# Label the axes
ax.set_xlabel('Dim 1')
ax.set_ylabel('Dim 2')
ax.set_zlabel('Dim 3')

# Display the plot
plt.title("a 3D Array (3x3x3)")
plt.show()
```
